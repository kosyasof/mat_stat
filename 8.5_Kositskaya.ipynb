```python
def kumaraswamy_logL(par, data):
    N = len(data)
    a,b = par
    logL = N*np.log(a*b) + (a-1)*np.sum(np.log(data))\
    + (b-1)*np.sum(np.log(1-np.power(data, a)))
    return logL
```

```python
def kumaraswamy_mle(data):
    res = opt.minimize(
    fun=lambda params, data: -kumaraswamy_logL(params, data),\
    x0=np.array([0.5, 0.5]), args=(data,), method='BFGS')
    a,b =res.x
    return a,b
```
