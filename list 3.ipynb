```python
from sympy import *
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sps
from scipy.special import gamma as Gamma
from IPython.display import display, Math, Latex
```

```python
x, s = symbols('x, s')
```

```python
def f1(x):
    return Gamma(x)


x = np.linspace(-3.5, 6, 512)
y1 = f1(x)

fig, ax = plt.subplots()
ax.plot(x, y1)
plt.xlim([-3.5, 6])
plt.ylim([-50, 50])
plt.grid()
plt.show()
```

```python
X = np.linspace(0, 4, 256)
g = f1(X)
print(' (s) ', min(g))
```

$E[\chi^2_k] = k\\
Var[\chi^2_k] = 2k\\
\chi^2(k_1) + \chi^2(k_2) \sim \chi^2(k_1 + k_2)\\
\chi^2_k \sim N(k, 2k),   \sigma = \sqrt{2k}$

```python
fig, ax = plt.subplots()
df = [2,3,5,10]
for i in df:
    x = np.linspace(0, 30, 10000)
    ax.plot(x, sps.chi2(i).pdf(x), label=f'{i}')
    ax.fill_between(x, sps.chi2(i).pdf(x), alpha=0.1)
plt.grid()
plt.legend()
plt.xlim([-0.1,10])
fig.set_figheight(5)
plt.ylim([0, 0.55])
fig.set_figwidth(10)
plt.show()
```

```python
display(Math(r'\chi^2_{0.1}(5) = ' + f'{round(sps.chi2.isf(0.1, 5), 4)}'))
display(Math(r'\chi^2_{0.01}(100) = ' + f'{round(sps.chi2.isf(0.01, 100), 4)}'))
display(Math(r'\chi^2_{0.95}(18) = ' + f'{round(sps.chi2.isf(0.95, 18), 4)}'))
display(Math(r'\chi^2_{0.99}(80) = ' + f'{round(sps.chi2.isf(0.99, 80), 4)}'))
```

```python
fig, ax = plt.subplots()
df = [2,3,5,10]
for i in df:
    x = np.linspace(-10, 10, 1000)
    ax.plot(x, sps.t(i).pdf(x))
plt.grid()
fig.set_figheight(5)
fig.set_figwidth(10)
plt.xlim([-5, 5])
plt.show()
```

```python
display(Math(r'P(t_{10} < 2.23) = '+ f'{sps.t(10).cdf(2.23)}'))
display(Math(r'P(t_{10} > -2.23) = '+ f'{sps.t(10).sf(-2.23)}'))
display(Math(r'P(|t_{10}| < 2.23) = '+ f'{sps.t(10).cdf(2.23) - sps.t(10).cdf(-2.23)}'))
display(Math(r'P(-1.31 < t_{10} < 3.17) = '+ f'{sps.t(10).cdf(3.17) - sps.t(10).cdf(-1.31)}'))
display(Math(r'P(t_{82} > 1.96) = '+ f'{sps.t(82).sf(1.96)}'))
```
