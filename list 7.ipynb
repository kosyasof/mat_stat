```python
Lambda= symbols('lambda', positive=True)
t, x = symbols('tau,x')
EX = integrate(x*Lambda*exp(-Lambda*(x-t)), (x, t , oo))
simplify(EX) # EX = 1/lambda + tau (сдвинуто от экспоненциального на тау)
```

```python
# руками можно показать через замену
tt = (x-t)
tt
```

```python
VARX = integrate((x-EX)**2*Lambda*exp(-Lambda*tt), (x, t , oo))
VARX
```
