```python
import numpy as np
from scipy.stats import binom, uniform, poisson
```

```python
#p1 - вероятность невыпадения, p2 - вероятность выпадания

def f1(p1, p2, n):
    E_XI = p2**2 * n
    Var_XI = E_XI * p1 * p2
    return f'мат ожидание равно:  {E_XI}, дисперсия усл. матожа равна:   {Var_XI}'
```

```python
f1(0.75, 1-0.75, 186)
```

```python
f1(0.55, 0.45, 50)
```

```python
def f2(p1, p2, n):
    E_XI = p2**2 * n
    return E_XI
```

```python
f2(0.35, 0.65, 118)
```

```python
#1
n = 186
p = 0.75
q = 1-p
fst = binom(n, p)
print("E(2го броска | 1го броска):", fst.mean() * p)
print("D(E(2го броска | 1го броска)):", fst.var() * p**2)
print("E(D(2го броска | 1го броска)):", fst.mean() * p * q)
```

```python
# X - ущерб от пожара, Y - число пожаров в год
def f4(EX, EY):
    VarY = EY
    VarX = EX**2
    ES = EX * EY
    VarS = VarX * EY + VarX * VarY
    return ES, VarS**(0.5)
```

```python
def f5(minu, maxu, EY):
    EX = (maxu + minu)/2
    VarX = (maxu - minu)**2/12
    VarY = EY
    ES = EX * EY
    VarS = EX**2 * EY + VarX * VarY
    return ES, VarS**(0.5)
```

```python
f5(0, 3.3, 12/10)
```

```python
f5(0, 4.8, 15/5)
```

```python
def f6(y1, y2, p1, p2, a, b):
    EY = y1*p1 + y2*p2
    EX = (a + b*y1)/2 * p1 + (a+b*y2)/2 * p2
    print(EX)
    EXY = y1*(a + b*y1)/2 * p1 + y2*(a+b*y2)/2 * p2
    Cov = EXY - EX*EY
    return EXY, Cov
```
